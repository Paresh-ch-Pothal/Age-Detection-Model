```python
import os
import cv2
from keras.models import load_model
```

```python
import numpy as np
import pandas as pd
```

```python
model = load_model('modelGender.h5')

# Define the expected input size of your model
IMG_SIZE = (224, 224)

# Load Haar Cascade for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Start camera
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # Face detection works better on grayscale
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

    for (x, y, w, h) in faces:
        # Extract face ROI
        face = frame[y:y+h, x:x+w]
        try:
            # Preprocess face
            face_resized = cv2.resize(face, IMG_SIZE)
            face_normalized = face_resized / 255.0
            input_face = np.expand_dims(face_normalized, axis=0)

            # Predict
            prediction = model.predict(input_face)[0][0]
            label = "Male" if prediction < 0.5 else "Female"

            # Draw green rectangle and label
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
            cv2.putText(frame, label, (x, y - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
        except Exception as e:
            print("Error processing face:", e)

    # Show frame
    cv2.imshow('Gender Prediction', frame)

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Cleanup
cap.release()
cv2.destroyAllWindows()
```

```python
def imagePrediction(imgPath):
    # Read and preprocess image
    img = cv2.imread(imgPath)
    img = cv2.resize(img, (224,224))
    img = img / 255.0 
    img = np.expand_dims(img, axis=0)  # shape: (1, 224, 224, 3)

    # Predict
    prediction = model.predict(img)
    if prediction >= 0.5:
        print("Female")
    else:
        print("male")
```
